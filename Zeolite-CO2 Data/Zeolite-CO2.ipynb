```python
# Function to search for DOI in all rows
def extract_doi(dataframe):
    for index, row in dataframe.iterrows():
        # Check if the row contains DOI information
        if 'DOI' in str(row.values):
            doi = (row.str.split(','))[0][1]
            return doi.strip('"')  # Remove quotes if present
    return None  # Return None if DOI not found

# Extract DOI
doi = extract_doi(df)
print(f"Extracted DOI: {doi}")
```
